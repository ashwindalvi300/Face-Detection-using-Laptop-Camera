```python
import cv2
import mediapipe as mp
```

```python
# Initialize the mediapipe face detection model
face_detection = mp.solutions.face_detection.FaceDetection(0.6)
```

```python
# Function for detecting faces in a frame
def detector(frame):
    count = 0
    height, width, channel = frame.shape

    # Convert frame from BGR to RGB as required by mediapipe
    imgRGB = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Process the frame for face detection
    result = face_detection.process(imgRGB)

    # If detections are found
    try:
        for count, detection in enumerate(result.detections):

            # Get the detection score and bounding box
            score = detection.score
            box = detection.location_data.relative_bounding_box

            # Convert bounding box from normalized coordinates to pixels
            x, y, w, h = int(box.xmin * width), int(box.ymin * height), int(box.width * width), int(box.height * height)
            score = str(round(score[0] * 100, 2))

            # Draw the bounding box around the detected face
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 0, 255), 2)
            cv2.rectangle(frame, (x, y), (x + w, y - 25), (0, 0, 255), -1)

            # Put the confidence score on the frame
            cv2.putText(frame, score, (x, y), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 255, 255), 2)

        count += 1
        print("Found", count, "faces!")

    except:
        pass

    return count, frame
```

```python
# Open the webcam (use 0 for the default camera)
cap = cv2.VideoCapture(0)
```

```python
# Loop to continuously capture frames from the camera
while True:
    ret, frame = cap.read()

    if not ret:
        break

    # Detect faces in the current frame
    count, output = detector(frame)

    # Display the number of faces detected on the frame
    cv2.putText(output, "Number of Faces: " + str(count), (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

    # Show the frame with the detected faces
    cv2.imshow("Face Detection", output)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```

```python
# Release the camera and close windows
cap.release()
cv2.destroyAllWindows()
```
